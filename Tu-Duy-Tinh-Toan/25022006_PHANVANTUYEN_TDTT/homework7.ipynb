```python
#w7a3
def bubblesort(a):
    l = len(a)
    for i in range(1, l - 1):
        if a[i] < a[i - 1]:
            a[i], a[i - 1] = a[i - 1], a[i]
    return a 

ds = list(map(int, input().split()))
print(bubblesort(ds))

def selectionsort(a):
    b = []
    for i in range(1, len(a) - 1):
        k = min(a[i - 1], a[i])
        k1 = min(k, a[i - 1])
        b.append(k1)
        for u in a:
            if u == k1:
                a.remove(u)
                break
    return b


```
