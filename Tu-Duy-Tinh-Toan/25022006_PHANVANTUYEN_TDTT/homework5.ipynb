```python
#w5a4
def shhcheck(a):
    s = 0
    for i in range(1, a):
        if a % i == 0:
            s += i
    if a == s:
        return True
    return False

n = int(input())
if shhcheck(n) == True:
    print('la so hoan hao')
else:
    print('khong la so hoan hao')
```

```python
#w5a6
def giaithua(n):
    e = 1 
    for i in range(1, n + 1):
        e = e * i
    return e
n = int(input())
print(giaithua(n))
```

```python
#w5a7
def tinh(num1, num2, kieu):
    tinh = 0
    if kieu == '+':
        tinh = num1 + num2
    elif kieu == '-':
        tinh = num1 - num2
    elif kieu == '/':
        tinh = num1 / num2
    else:
        tinh = num1 * num2
    return tinh


num1 = float(input())
num2 = float(input())
kieu = input()

print(round(tinh(num1, num2, kieu), 2))
```

```python
#w5a9
def sumscs(n):
    sum = 0
    while n > 0:
        sum += n % 10
        n //= 10
    return sum
n = int(input())
print(sumscs(n))
```
