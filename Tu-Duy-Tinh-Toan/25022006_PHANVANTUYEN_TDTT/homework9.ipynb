```python
#w9a1 
def merge(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    arr1 = arr[:mid]
    arr2 = arr[mid:]
    
    ans = [0] * len(arr)
    arr1 = merge(arr1)
    arr2 = merge(arr2)

    i = j = k = 0

    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            ans[k] = arr1[i]
            i += 1
            k += 1
        else:
            ans[k] = arr2[j]
            j += 1
            k += 1
        
    while i < len(arr1):
        ans[k] = arr1[i]
        i += 1
        k += 1
    while j < len(arr2):
        ans[k] = arr2[j]
        j += 1
        k += 1
    
    return ans

a_list = [1,3,4,5,6,66,33,54,5,5,4,-1]
print(merge(a_list))
    
    
```

```python
#w9a4
def main(a, b):
    cnt_a = 0
    for i in range(len(a)):
        if ord('A') <= ord(a[i]) <= ord('Z'): cnt_a += 1

    j = 0
    cnt_b = 0
    pos = [] #tuple of b[i], index b[i], index a[j]
    for i in range(len(b)):
        while j < len(a) and ord('a') <= ord(a[j]) <= ord('z'):
            j += 1
        
        if j < len(a) and b[i] == a[j]:
            cnt_b += 1
            pos.append((b[i], i, j))
            j += 1
        elif i == 0 and len(pos) == 0:
            pos.append((-1, -1, -1))
    
    pos.append((-1, len(b), len(a)))
    
        
    if cnt_a != cnt_b: return 0

    a = a.upper()
    for i in range(1, len(pos)):
        left = pos[i-1][2] + 1
        right = pos[i][2] - 1
        for j in range(pos[i-1][1] + 1, pos[i][1]):
            check = 0

            for k in range(left, right + 1):
                if a[k] == b[j]:
                    check = 1
                    left = k + 1
                    break
            
            if check == 0: return 0
    
    return 1
            


while 1:
    try:
        t = int(input())
        while t > 0:
            a = input()
            b = input()
            if (main(a, b) == 0): print('NO')
            else: print('YES')
            t -= 1
        
        print('')

    
    except EOFError:
        break
```

```python
#w9a6
def next_permutation(a):
    i = len(a) - 2
    while i >= 0 and a[i] >= a[i+1]: i -= 1

    if i == -1:
        a.reverse()
        return a

    j = len(a) - 1
    while j > i and a[j] <= a[i]: j -= 1

    a[i], a[j] = a[j], a[i]

    a[i+1:] = reversed(a[i+1:])

    return a

n = int(input())
a = list(map(int, input().split()))
res = next_permutation(a)
for i in res: print(i, end = ' ')
print('')
```

```python
#w9a7
def canPlaceFlowers(a, k):
    if len(a) == 1 and a[0] == 0:
        k -= 1
    
    if len(a) == 1: return (k <= 0)

    for i in range(len(a)):
        if 0 < i < len(a) - 1 and a[i-1] == a[i] == a[i+1] == 0:
            a[i] = 1; k -= 1
        elif i == 0 and a[i] == a[i+1] == 0:
            a[i] = 1; k -= 1
        elif i == len(a) - 1 and a[i-1] == a[i] == 0:
            a[i] = 1; k -= 1
    
    return (k <= 0)

a = list(map(int, input().split()))
    k = int(input())
    print(canPlaceFlowers(a, k))
```

```python
#w9a8
move = ['L', 'R', 'U', 'D']
move_x = [0, 0, 1, -1]
move_y = [-1, 1, 0, 0]

def is_path_crossing(moves):
    trace = [(0, 0)]
    curpos = (0, 0)
    for i in range(len(moves)):
        z = move.index(moves[i])
        curpos = (curpos[0] + move_x[z], curpos[1] + move_y[z])
        if curpos in trace: return True
        trace.append(curpos)
    
    return False



moves = input()
print(is_path_crossing(moves))
  
```

```python
#w9a9

def median(nums):
    def max_idx(nums):
        max = 1
        for k in nums:
            if k > max:
                max = k
        return max
    
    for k in range(2):
        value_remove = max_idx(nums)
        nums.remove(value_remove)
    return nums[-1]

data_input = list(map(int, input().split()))
print(median(data_input))

```
