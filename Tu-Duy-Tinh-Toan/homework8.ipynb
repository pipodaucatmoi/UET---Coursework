```python
#w7a27 - bai toan day ngoac dung

ngoac = input()

def check_ngoac(data_ngoac):
    stack = []
    key_dong = {')' : '(' , ']' : '[', '}' : '{'}

    for k in data_ngoac:
        if k not in key_dong:
            stack.append(k)
        else:
            if len(stack) > 0:
                key1 = stack.pop()
            else:
                key1 = '@'
            if key1 != key_dong[k]:
                return False
    if len(stack) == 0:
        return True


print(check_ngoac(ngoac))
```

```python
#w7a28 

n = int(input())
data_list = []

def check_snt(n):
    if n < 2:
        return False
    if n == 2:
        return True
    for i in range(2, int(n ** 0.5)):
        if n % i == 0:
            return False
            break
    return True

for i in range(1, n):
    if check_snt(i) == False:
        data_list.append(i)

count = {data_list[0] : 1}
ans_list = []

for i in range(1, len(data_list)):
    if data_list[i] - data_list[i - 1] == 1:
        count[data_list[i]] = count[data_list[i - 1]] + 1
        ans_list.append(data_list[i])

print(ans_list[0])



```
