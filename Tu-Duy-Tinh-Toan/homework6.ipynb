```python
#w6a2
def tinhtong(x, i):
    sum = 0
    for k in x[:i+1]:
        sum += k
    return sum

A = list(map(int, input().split()))
n = len(A)

for i in range (0, n):
    print(tinhtong(A, i), end = ' ')
```

```python
#w6a5

ds = list(map(int, input().split()))
dic_now = {'positives' : 0, 'negatives' : 0, 'zeros' : 0}

for k in ds:
    if k == 0:
        dic_now['zeros'] += 1
    elif k > 0:
        dic_now['positives'] += 1
    else:
        dic_now['negatives'] += 1

print(f'positives : {dic_now['positives']}, negatives : {dic_now['negatives']}, zeros = {dic_now['zeros']}')
```

```python
#w6a9

def anhxa(dic1, dic2):
    sum_dic = {}
    for k in dic1:
        if k in dic2:
            sum_dic[k] = dic1[k] + dic2[k]
        else:
            sum_dic[k] = dic1[k]
    
    for k in dic2:
        if k not in dic1:
            sum_dic[k] = dic2[k]
    
    return sum_dic

d1 = {"a": 1, "b": 2}
d2 = {"b": 3, "c": 4}
print(anhxa(d1, d2))
```
